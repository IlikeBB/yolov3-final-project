```python
import xml.etree.ElementTree as ET
import pickle
import os
import sys
from os import listdir, getcwd
from os.path import join
image_set='train'

classes = ['aquarium','bottle','bowl','box','bucket','plastic_bag','plate','styrofoam',
         'tire','toilet','tub','washing_machine','water_tower']
```

```python
def convert(size, box):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)
```

```python
def convert_annotation(image_id):
    
    in_file = open('../data/Annotations/%s/%s.xml'%(image_set,image_id))
    
    out_file = open('../data/Imagetxt/%s/%s.txt'%(image_set,image_id),'w')#生成txt格式文件
    tree=ET.parse(in_file)
    root = tree.getroot()
    size = root.find('size')  
    w = int(size.find('width').text)
    h = int(size.find('height').text)

    for obj in root.iter('object'):
        cls = obj.find('name').text
        if cls not in classes :
            continue
        cls_id = classes.index(cls)
        print('pass')
        xmlbox = obj.find('bndbox')   
        b = (float(xmlbox.find('xmin').text), float(xmlbox.find('xmax').text), float(xmlbox.find('ymin').text), float(xmlbox.find('ymax').text))
        bb = convert((w,h), b)
        out_file.write(",".join([str(a) for a in bb]) + ","+str(cls_id)+'\n')


if image_set=='train':
    image_ids_train = open('../data/labels/label_%s.txt'%(image_set)).read().strip().split()#list格式只有000000 000001
    list_file_train = open('../data/dataset/train.txt', 'w')
    for image_id in image_ids_train:
        list_file_train.write('../data/%s/%s.jpg\n'%(image_set,image_id))  
        convert_annotation(image_id)   
    list_file_train.close()      #只生成训练集，自己根据自己情况决定
    
elif image_set=='test':
    image_ids_test = open('../data/labels/label_%s.txt'%(image_set)).read().strip().split()
    list_file_test = open('../data/dataset/test.txt', 'w')
    for image_id in image_ids_train:
        list_file_train.write('data/%s/%s.jpg\n'%(image_set,image_id))  
        convert_annotation(image_id)   
    list_file_train.close()      #只生成训练集，自己根据自己情况决定
```
