```python
import xml.etree.ElementTree as ET
import pickle
import os
from os import listdir, getcwd
from os.path import join

sets=[('2007', 'test')]

classes = ["aquarium", "bottle", "bowl", "box", "bucket", "plastic_bag", 
           "plate", "styrofoam", "tire", "toilet", "tub", "washing_machine", "water_tower"]
```

```python
def convert_annotation(year, image_id,image_set):
    in_file = open('../data/Annotations/%s/%s.xml'%(image_set,image_id))
    
    out_file = open('../data/Imagetxt/%s/%s.txt'%(image_set,image_id),"w")
    tree=ET.parse(in_file)
    root = tree.getroot()
    size = root.find('size')
    w = int(size.find('width').text)
    h = int(size.find('height').text)
    for obj in root.iter('object'):
        cls = obj.find('name').text
        cls_id = classes.index(cls)
        xmlbox = obj.find('bndbox')
        b = (int(xmlbox.find('xmin').text), int(xmlbox.find('xmax').text), int(xmlbox.find('ymin').text), int(xmlbox.find('ymax').text))
#         bb = convert((w,h), b)
#         print(",".join([str(a) for a in b]) +","+str(cls_id)+' ')
        out_file.write(",".join([str(a) for a in b]) +","+str(cls_id)+' ')
#         out_file.write(",".join([str(a) for a in bb])+' ')

wd = getcwd()


for year, image_set in sets:
    
    image_ids = open('../data/labels/label_%s.txt'%(image_set)).read().strip().split()
#     list_file = open('%s_%s.txt'%(year, image_set), 'w')
    for image_id in image_ids:

#         list_file.write('VOC%s/ImageJPEG/%s/%s.jpg\n'%(year, image_set,image_id))
        convert_annotation(year, image_id,image_set)
        
        out_file_txt=open('../data/Imagetxt/%s/%s.txt'%(image_set,image_id), 'r')
        content=out_file_txt.read()
        out_file_txt.close()
        
        out_file_txt=open('../data/Imagetxt/%s/%s.txt'%(image_set,image_id), 'w')
        content_add=('data/ImageJPEG/%s/' % (image_set)+str(image_id)+'.jpg '+content[:-1])
#         content_add=(content[:-1])
        out_file_txt.write(content_add)
#         print(content_add)
        out_file_txt.close()
#     list_file.close()
```
